# Dashboard

We r not doing data mining, investigation or analysis. We r creating pretty and powerful tool for someone to do data investigation through worksheets we have created.

Instead of dimensions and measures, we have **Sheets** available in dashboard page. Also u can add objects - images, text, horizontal, vertical layouts. Can make objects floating or tiled.

## Horizontal and vertical containers in objects

Horizontal and vertical containers can help to split your dashboard into segments.

Adding filter on dashboard

## Making Dashboard Interactive

Allow user of dashboard to drill into certain parts of dashboard by just simply clicking on it.

e.g. if u want to just see france in map then the customer scatterplot will also change accordingly if dashboard interactive.
